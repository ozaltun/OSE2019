```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('parallel_pi.txt', delimiter='\t')
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,5))
for i in data['Max Thread '].unique().tolist():
    temp_data = data[data['Max Thread '] == i]
    ax.scatter(temp_data['Size of array'], temp_data['Parallel time '], label='Thread: {}'.format(i))

ax.set_xlabel('Number of points')
ax.set_ylabel('Time (in seconds)')

fig.tight_layout()
fig.legend()
plt.show()
```
