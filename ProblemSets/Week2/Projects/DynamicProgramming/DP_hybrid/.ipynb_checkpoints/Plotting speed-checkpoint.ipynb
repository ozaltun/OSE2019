```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
%matplotlib notebook
```

```python
def plot_data(root):
    data = np.loadtxt(root) 
    # Column 1 is points, column 2 is number of processes, column 3 is number of threads, column 4 is error
    fig = plt.figure()
    fig.tight_layout()
    ax = fig.add_subplot(111, projection='3d')
    dem3d=ax.plot_surface(data[:,1].reshape(4, 4), data[:,2].reshape(4, 4), data[:,3].reshape(4,4))
    ax.set_xlabel('Number of units')
    ax.set_ylabel('Number of threads')
    ax.set_zlabel('Time (s)')
    ax.set_title('Number of grid points:{}'.format(root.split('_')[0]))
```

```python
plot_data('1000_test.txt')
```

```python
plot_data('5000_test.txt')
```

```python
plot_data('10000_test.txt')
```

```python
fig = plt.figure()
fig.tight_layout()
ax = fig.add_subplot(111, projection='3d')


for i in ['1000_test.txt','5000_test.txt','10000_test.txt']:
    data = np.loadtxt(i) 
    # Column 1 is points, column 2 is number of processes, column 3 is number of threads, column 4 is error

    ax.scatter(data[:,1].reshape(4, 4), data[:,2].reshape(4, 4), data[:,3].reshape(4,4), label='grid points:{}'.format(i.split('_')[0]))
    ax.set_xlabel('Number of units')
    ax.set_ylabel('Number of threads')
    ax.set_zlabel('Time (s)')
ax.legend()
    
plt.show()
```
