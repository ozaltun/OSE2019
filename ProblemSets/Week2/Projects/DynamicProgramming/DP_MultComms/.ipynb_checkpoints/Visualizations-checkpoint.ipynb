```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
%matplotlib notebook
```

```python
def plot_data(root):
    data = np.loadtxt(root) 
    # Column 1 is points, column 2 is number of processes, column 3 is number of threads, column 4 is error
    fig = plt.figure()
    fig.tight_layout()
    ax = fig.add_subplot(111, projection='3d')
    dem3d=ax.scatter(data[:,1].reshape(3, 4), data[:,2].reshape(3, 4), data[:,3].reshape(3,4))
    ax.set_xlabel('Number of units')
    ax.set_ylabel('Number of threads')
    ax.set_zlabel('Time (s)')
    ax.set_title('Number of grid points:{}'.format(36000))
```

```python
plot_data('hybrid_many_settings_DP.txt')
```
