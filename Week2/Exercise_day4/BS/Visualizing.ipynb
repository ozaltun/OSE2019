```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = open('8_threads_hybrid.txt').readlines()
data = [i for i in data if 'Time' in i]
data = [float(i[-8:-1]) for i in data]
```

```python
fig = plt.subplots(figsize=(10,5))
for i, time in enumerate(data):
    plt.scatter(2**i, time, label = '{} processes'.format(2**i))
plt.legend()
plt.xlabel('Number of processes')
plt.ylabel('Computation time in seconds')
plt.title('Time to compute BS.cpp given 8 threads of Open MP and different number of \
MPI processes')
plt.show()
```
